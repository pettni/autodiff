```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(
    font_scale=2,
    palette=sns.color_palette()
)
```

```python
df = pd.read_csv('data', names=["Library", "Test", "Size", "Setup", "Calc"], sep='\s+')

for size in df['Size'].unique():

    plt.figure(figsize=(30,16))
    ax = sns.barplot(data=df.loc[df['Size'] == size], x="Test", y="Calc", hue="Library")
    ax.set(title="Size {}".format(size), yscale="log", xlabel="", ylabel="Time", ylim=(1e-1, 1e7))
    ax.legend(loc="upper center", shadow=False, ncol = len(df["Library"].unique()))

    plt.show()
```
